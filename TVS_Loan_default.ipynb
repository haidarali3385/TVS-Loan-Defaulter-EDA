```python
df.rename({'V1':"Customer ID"},axis=1,inplace=True)
df.rename({'V2':"1M Bounce"},axis=1,inplace=True)
df.rename({'V3':"12M Bounce"},axis=1,inplace=True)
df.rename({'V4':"TVS Tenure"},axis=1,inplace=True)
df.rename({'V5':"No of Bouncing"},axis=1,inplace=True)
df.rename({'V6':"EMI"},axis=1,inplace=True)
df.rename({'V7':"Loan Amount"},axis=1,inplace=True)
df.rename({'V8':"Loan Tenure"},axis=1,inplace=True)
df.rename({'V9':"Dealer Code"},axis=1,inplace=True)
df.rename({'V10':"Two Whealer Code"},axis=1,inplace=True)
df.rename({'V11':"No Advanced EMI"},axis=1,inplace=True)
df.rename({'V12':"ROI"},axis=1,inplace=True)
df.rename({'V13':"Gender"},axis=1,inplace=True)
df.rename({'V14':"Occupation"},axis=1,inplace=True)
df.rename({'V15':"Resident Type"},axis=1,inplace=True)
df.rename({'V16':"DOB"},axis=1,inplace=True)
df.rename({'V17':"Age"},axis=1,inplace=True)
df.rename({'V18':"No Of Loans"},axis=1,inplace=True)
df.rename({'V19':"No Of Secure Loan"},axis=1,inplace=True)
df.rename({'V20':"No Of Unsecure Loan"},axis=1,inplace=True)
df.rename({'V21':"Live Loan Sanction Amount"},axis=1,inplace=True)
df.rename({'V22':"Last 3M New Loans"},axis=1,inplace=True)
df.rename({'V23':"Amount Live Secure Loan"},axis=1,inplace=True)
df.rename({'V24':"Amount Live Unsecure Loan"},axis=1,inplace=True)
df.rename({'V25':"Amount Of Any Two Whealer"},axis=1,inplace=True)
df.rename({'V26':"Last PL Tacken Time In Month"},axis=1,inplace=True)
df.rename({'V27':"Fisrt CD Tacken Time In Month"},axis=1,inplace=True)
df.rename({'V28':"No Of DPD 30D in 6M"},axis=1,inplace=True)
df.rename({'V29':"No Of DPD 60D in 6M"},axis=1,inplace=True)
df.rename({'V30':"No Of DPD 90D in 3M"},axis=1,inplace=True)
df.rename({'V31':"Tier"},axis=1,inplace=True)
df.rename({'V32':"Target"},axis=1,inplace=True)
```
